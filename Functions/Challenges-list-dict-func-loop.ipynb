```python
# #-------------------> DIFICULDADE 1 <---------------------------
# # Um algoritmo capaz de fazer uma contagem de 1 ao 9.
# i=0
# while i<9:
#   print(i+1)
#   i += 1
  
# # Recebe como entrada o raio do círculo em centímetros e vai mostrar no terminal pro usuário o resultado da fórmula para calcular a área.

# raio = int(input('Digite o raio da circunferencia em cm: '))
# area =  3.14 * raio**2
# print('A area desta circunferencia e', area)

# # Desenvolva um programa que leia um número inteiro qualquer e que apresente o número informado, seguido do seu antecessor e do seu sucessor

# num = int(input('Digite um numero: '))
# print(f'O numero digitado foi {num}. O numero que se antecede a este numero e : {num - 1}. O numero que sucede este numero e {num + 1}. ')

# # Desenvolva um programa que leia um número inteiro qualquer e que informe se este número é par ou impar

# num = int(input('Digite um numero: '))
# if num % 2 == 0:
#   print(f'O numero digitado foi {num} e este numero e par.')
# else:
#   print(f'O numero digitado foi {num} e este numero e impar.')

# #Um algoritmo que recebe uma temperatura em graus Celsius e retorna a temperatura em graus Fahrenheit.

# temp_c = float(input('Digite a temperature em Celsius: '))
# temp_f = temp_c * 9/5 + 32 
# print(f'{temp_c} graus Celsius e equivalente a {temp_f} graus Fahrenheit.')

# # Código capaz de criar uma lista que adiciona 5 número reais e imprimir a lista na ordem inversa.

# nums = []
# for i in range(5):
#   n = input(f'Digite {i+1}o numero: ')
#   nums.append(n)

# print(nums[::-1])


# #-------------------> DIFICULDADE 2 <---------------------------
# # Desenvolva um programa que tenha uma função que verifique se um número inteiro qualquer é par ou impar

# def par_impar(n):
#   if n % 2 == 0:
#     print('E par!')
#   else:
#     print('E impar!')

# num = int(input('Digite um numero:'))
# par_impar(num)

# # Desenvolva um programa que leia quatro notas e que apresente a média final

# notas = []

# for i in range(4):
#   nota = float(input(f'Digite sua {i+1} nota: '))
#   notas.append(nota)

# media = sum(notas)/len(notas)
# print(f'A sua media foi {media}.')


# # Considere o seguinte dicionário: {'m1': {'m2': 'Olá Mundo'}}. Carregue e apresente a mensagem "Olá Mundo" contida no dicionário

# dic = {'m1': {'m2': 'Olá Mundo'}}
# print(dic['m1']['m2'])

# # Algoritmo que recebe 5 valores de entrada, no final irá printar duas listas, a primeira com os valores informados e a segunda com todos os valores ao quadrado.

# nums   = []
# nums_2 = []
# for i in range(5):
#   num = float(input(f'Digite o {i+1}o numero: '))
#   nums.append(num)
#   nums_2.append(num**2)
  
# print(nums)
# print(nums_2)

# # Programa que recebe como entrada 6 valores inteiros, no final ele exibe uma lista com os 6 valores informados e qual o seu maior elemento.

# nums = []
# for i in range(6):
#   num = int(input(f'Digite o {i+1}o numero inteiro: '))
#   nums.append(num)

# print(nums)
# print(max(nums))

# # Algoritmo que solicita três lados de um triângulo, com esses dados o nosso programa deve informar a nomenclatura dos triângulos.
# # Triângulo Equilátero: três lados iguais;
# # Triângulo Isósceles: quaisquer dois lados iguais;
# # Triângulo Escaleno: três lados diferentes;

# lados = []
# for i in range(3):
#   lado = int(input(f'Digite o {i+1}o lado da figura geometrica: '))
#   lados.append(lado)

# if lados.count(lados[0]) > 2:
#   print('Triangulo Equilatero')
# elif lados.count(lados[0]) > 1:
#   print('Triangulo Isosceles')
# else:
#   print('Triangulo Escaleno')

# # Programa capaz de solicitar 4 valores para representar a nota do aluno, em seguida armazenas os valores em uma lista, no fim mostrar as notas e a média.

# notas  = []
# for i in range(4):
#   nota = float(input(f'Digite sua {i+1} nota: '))
#   notas.append(nota)

# print(notas)
# print('A media de notas e', sum(notas)/len(notas))

  
# #-------------------> DIFICULDADE 3 <---------------------------

# # Desenvolva um programa que leia um número inteiro qualquer e que apresente o número informado com duas casas decimais

# num = int(input('Digite um numero inteiro: '))
# print(float(num))

# # Programa consiste em uma estrutura de repetição que vai fazer a leitura de vários números inteiros e impares, quando a lista chegar a 5 elementos deve parar com a leitura.

# lista = []
# while True:
#   num = int(input('Digite um numero: '))
#   if num %2 != 0:
#     lista.append(num)
#   if len(lista) == 5:
#     print(lista)
#     break

# # Programa capaz de lidar com uma lista de 5 número inteiros e imprimir cada número juntamente com a sua posição na lista

# lista = []
# for i in range(5):
#   num = int(input(f'Digite o {i+1}o numero: '))
#   lista.append(num)

# for i in range(len(lista)):
#   print(f'O numero na {i+1}a posicao da lista e {lista[i]}')



# #-------------------> DIFICULDADE 4 <---------------------------
# # Programa que simula uma agenda de contatos e armazena-os em um dicionário, na questão temos 3 contatos e cada contato pode ter vários números.

# #solution1
# nums = []
# nomes= []
# for i in range(3):
#   x = 1
#   nome = input(f'Digite o nome do {i+1}o contato da sua agenda: ')
#   nomes.append(nome)
#   num  = input(f'Digite o {x}o numero deste contato: ')
#   nums.append(num)

#   while True:
#     cont = (input('Este contato possue outro numero? Digite S para sim or N para nao: ')).upper()
#     if cont == 'S':
#       nomes.append(nome + '2')
#       num  = input(f'Digite o {x+1} numero deste contato: ')
#       nums.append(num)
#       x += 1
#     else:
#       break
  
# print(dict(zip(nomes,nums)))

# #solution2 - Muito melhor haha
# contato = {}
# for i in range(3):
#   nums = []
#   x = 1
#   nome = input(f'Digite o nome do {i+1}o contato da sua agenda: ')
#   num  = input(f'Digite o {x}o numero deste contato: ')
#   nums.append(num)
#   contato[nome] = nums

#   while True:
#     cont = (input('Este contato possue outro numero? Digite S para sim or N para nao: ')).upper()
#     if cont == 'S':
#       num  = input(f'Digite o {x+1} numero deste contato: ')
#       contato[nome].append(num)
#       x += 1
#     else:
#       break

# print(contato)      

# #Solution 3 A melhor versao! \o/

# contato = {}
# for i in range(3):
#   nums = []
#   nome = input(f'Digite o nome do {i+1}o contato da sua agenda: ')
#   rep  = int(input('Quantos numeros tem este contato? '))
#   for r in range(rep):
#     num  = input(f'Digite o {r+1}o numero deste contato: ')
#     nums.append(num)
#   contato[nome] = nums
# print(contato)  


#Solution4 Mais realistica

# def add_contact(agenda):
#   nome = input('Digite o nome do contato a ser adicionado a sua agenda: ')
#   nums = []
#   while True:
#     num  = int(input('Digite o numero deste contato: '))
#     nums.append(num)
#     print('Contato adicionado.')
#     adicional = (input('Este contato tem mais de um numero? Digite S para sim ou N para nao: ')).upper()
#     if adicional == 'S':
#       continue
#     elif adicional == 'N':
#       agenda[nome] = nums
#       break
#     else:
#       print('Opcao invalida. Vamos tentar mais uma vez?')
#       continue


# def search_contact(agenda):
#   search = input('Search Contact: ')
#   find   = agenda.get(search, 0)
#   if find == 0:
#     print('Contato nao encontrado')
#     while True:
#       add = (input('Voce gostaria de adicionar este contato a sua agenda? Digite S para sim ou N para nao: ')).upper()
#       if add == 'S':
#         add_contact(agenda)
#         print('Sua agenda foi atualizada')
#         print(agenda)
#         break  
#       elif add == 'N':
#         print('Sua agenda nao sofreu modificacoes')
#         print(agenda)
#         break
#       else:
#         print('Opcao invalida. Vamos tentar mais uma vez?')
#   else:
#     print(f'''
#           Contato : {search} 
#           Numero  : {agenda[search]}''')


# agenda = {'Samsung voicemail' : {'Home Phone' : 699 ,
#           'Emergency' : 911}
# for i in range(3):
#   start = search_contact(agenda)

#solucao 5 criando dicionario dentro de dicionario:

def add_contact(agenda):
  name = input('Contact Name: ')
  print("Tip! If contact doesn't have a specific phone or email, Type 'None' to continue")
  num_res      = {'Home Phone' : input('Home Phone: ')}
  num_work     = {'Work Phone' : input('Work Phone: ')}
  num_cel      = {'Cel Phone'  : input('Cel  Phone: ')}
  email        = {'Email'      : input('Email: ')}
  agenda[name] = num_res,num_work,num_cel,email
  print('Contact was succesfully added!')
  while True:
    print('''Add new Contact       : Press 1
Return to Main Menu   : Press 2
''')
    try:
        do = int(input())
        if do < 1 or do > 2:
          print('Invalid Option')
          continue
        elif do == 1:
          add_contact(agenda)
          break
        elif do == 2:
          menu_contact(agenda)
          break
    except:
        print("Invalid Option")

  
    

def search_contact(search):
  find   = agenda.get(search, 0)
  if find == 0:
    print('Contact not found.\n')
    menu_contact(agenda)
  else:
    print(f'Contact: {search}')
    print(f'{agenda[search]}\n')
    # give the options to change contact from here later
    return 'found'



def open_agenda(agenda):
  print('      CONTACT LIST')
  for name in agenda:
    print(f'Contact name: {name} ')
    print(f'{agenda[name]}\n')
    
  


def delete_contact(name,agenda):
  check = search_contact(name)
  if check == 'found':
    while True:
      try:
        del_confirm = int(input(f'''Are you sure you want to delete the contact information of {name}?
        Yes : Press 1
        No  : Press 2
        '''))
        if del_confirm !=1 and del_confirm !=2:
          continue
        elif del_confirm == 1:
          agenda.pop(name)
          print(f'Contact {name} was succesfully deleted')
          menu_contact(agenda)
        else:
          menu_contact(agenda)
      except:
        print('Invalid Option')


def update_contact(name,agenda):
  check = search_contact(name)
  if check == 'found':
    while True:
      update = int(input('''To update:
        Home Phone  : Press 1
        Work Phone  : Press 2
        Cel Phone   : Press 3
        Email       : Press 4
        '''))
      if update == 1:
        new = input('Home Phone: ')
        agenda[name].update({'Home Phone' : new})
        break
            
      elif update == 2:
        new = input('Work Phone: ')
        agenda[name].update({'Work Phone' : new})
        break
                  
      elif update == 3:
        new = input('Cel Phone')
        agenda[name].update({'Cel Phone' : new})
        break
                  
      elif update == 4:
        new = input('Email: ')
        agenda[name].update({'Email' : new})
        break
  
      else:
        print('Invalid Option.')
        continue
  
  print('Contact was sucessfully updated')    
  open_contact(agenda)
    

def menu_contact(agenda):
  while True:
    open_agenda(agenda)
    print('          MENU')
    print('''Search Contact : Press 1
Add    Contact : Press 2
Close Agenda   : Press 3
    ''')
    try:
      action = int(input())
      if action < 1 or action > 3:
        print('Invalid Option')
        continue
      elif action == 1:
        find = input('Search Contact: ')
        search_return_contact(find)
        break
      elif action == 2:
        add_contact(agenda)
        break
      elif action == 3:
        break
    except:
      print("Invalid Option")
    




def search_return_contact(search):
  find   = agenda.get(search, 0)
  if find == 0:
    print('Contact not found.\n')
    while True:
      print('''Add Contact          : Press 1
Search Other contact : Press 2
Return to Main Menu  : Press 3
''')
      try:
        do = int(input())
        if do < 1 or do > 3:
          print('Invalid Option')
          continue
        elif do == 1:
          add_contact(agenda)
          break
        elif do == 2:
          new_search = input('Search Contact: ')
          search_return_contact(new_search)
          break
        elif do == 3:
          menu_contact(agenda)
          break
      except:
        print("Invalid Option")
  else:
    print(f'Contact: {search}')
    print(f'{agenda[search]}\n')
    while True:
      print('''Update Contact       : Press 1
Delete Contact       : Press 2
Search Other contact : Press 3
Return to Main Menu  : Press 4''')
      try:
        do = int(input())
        if do < 1 or do > 4:
          print('Invalid Option')
          continue
        elif do == 1:
          update_contact(search,agenda)
          break
        elif do == 2:
          delete_contact(search,agenda)
          break
        elif do == 3:
          new_search = input('Search Contact: ')
          search_return_contact(new_search)
          break
        elif do == 4:
          menu_contact(agenda)
          break
      except:
        print("Invalid Option")
   

  
agenda = {'Samsung voicemail' : {'Home Phone': 699, 'Work Phone': None, 'Cel Phone': None, 'Email': None},
          'Emergency' : {'Home Phone': 190, 'Work Phone': None, 'Cel Phone': None, 'Email': None}}
menu_contact(agenda)





  
  # # Um programa que le um dicionario com as disciplinas (chaves) e a nota (elemento), o código e capaz de dizer qual foi a menor nota.

# def get_key(elemento, dict):
#   for key,value in dict.items():
#     if elemento == value:
#       return key

# contato = {}
# n = int(input('Quantas disciplinas voce cursou esse semestre? '))
# for i in range(n):
#   disc = input(f'Digite a {i+1}a disciplina: ')
#   nota = int(input(f'Digite sua nota final na disciplina {disc}: '))
#   contato[disc] = nota
# menor_v = min(contato.values())  
# menor_k = get_key(menor_v,contato)
# print(f'A menor nota que voce tirou foi {menor_v} em {menor_k}.')




# # Em uma competição de salto em distância cada atleta tem direito a cinco saltos. No final da série de saltos de cada atleta, o melhor e o pior resultados são eliminados. O seu resultado fica sendo a média dos três valores restantes. Você deve fazer um programa que receba o nome e as cinco distâncias alcançadas pelo atleta em seus saltos e depois informe a média dos saltos conforme a descrição acima informada (retirar o melhor e o pior salto e depois calcular a média). Faça uso de uma lista para armazenar os saltos. Os saltos são informados na ordem da execução, portanto não são ordenados. O programa deve ser encerrado quando não for informado o nome do atleta. A saída do programa deve ser conforme o exemplo abaixo:
# # Atleta: Rodrigo Curvêllo

# # Primeiro Salto: 6.5 m
# # Segundo Salto: 6.1 m
# # Terceiro Salto: 6.2 m
# # Quarto Salto: 5.4 m
# # Quinto Salto: 5.3 m

# # Melhor salto:  6.5 m
# # Pior salto: 5.3 m
# # Média dos demais saltos: 5.9 m

# # Resultado final:
# # Rodrigo Curvêllo: 5.9 m


# def media_atleta(nome):
#   count = ['Primeiro','Segundo','Terceiro','Quarto','Quinto']
#   saltos = []
#   for i in range(5):
#     salto = float(input(f'Digite a altura do {count[i]} salto: '))
#     saltos.append(salto)
    
#   print('\n')
  
#   print(f'Atleta: {nome} \n')
  
#   for x in range(len(saltos)):
#     print(f'{count[x]} Salto: {saltos[x]} m')
  
#   print('\n')
  
#   print(f'Melhor salto: {max(saltos)} m')
#   print(f'Pior salto: {min(saltos)} m')
  
#   saltos.remove(min(saltos))
#   saltos.remove(max(saltos))
#   media = round((sum(saltos)/len(saltos)),2)
#   print(f'Media dos demais saltos: {media} m ')
  
#   print('\n')
  
#   print(f'''Resultado final:
# {nome} : {media} m.''')
 
# nome_atl = input('Digite o nome completo do atleta: ')
# media_atleta(nome_atl)


#-------------------> DIFICULDADE 5 <---------------------------
# Desenvolva um jogo de acerte o número, onde o computador escolhe um número inteiro aleatório de 0 a 10, e o usuário tem 5 tentativas para adivinhar o número
#a) Implemente um sistema de pontuação com o seguinte comportamento: se o usuário adivinhar o número na primeira tentativa, receberá a pontuação máxima (ex. 100 pontos); se o usuário adivinhar o número na última tentativa, receberá a pontuação mínima (ex. 10 pontos); se o usuário não acertar o número, não receberá nenhum ponto.
#b): Implemente um controle de erros. Caso o jogador digite um número fora da faixa permitida ou caracteres não numéricos, o sistema deve notificar o jogador e solicitar o input correto.
#c): Implemente a opção de o usuário iniciar uma nova partida. Ao finalizar uma rodada, após o resultado final, o jogo deve perguntar se o jogador quer iniciar uma nova partida e, em caso negativo, encerrar a aplicação.

# import random as r

# def new_game(p):
#   while True:
#     nova = (input('Voce gostaria de jogar de novo? Digite 1 continuar a jogar ou 2 para terminar o jogo '))
#     if nova == '1':
#       count = 0
#       n = r.randrange(0,10)
#       return count, n
#     elif nova == '2':
#       count = 6
#       n = 0
#       print(f'Voce fez no total {p} pontos nesse jogo.')
#       print('----Fim de Jogo----')
#       return count, n
#     else:
#       print('Opcao invalida. Por favor digite 1 para continuar ou 2 para terminar o jogo')
#       continue

# def tip(tentativa, acerto):
#   if tentativa < acerto:
#     print('O numero digitado e menor que o numero que a gente escolheu. Vamos tentar mais uma vez?')
#   else: 
#     print('O numero digitado e maior que o numero que a gente escolheu. Vamos tentar mais uma vez? ')

# def start_game(num):
#   pts = [100,77.5,55,32.5,10]
#   i = 0
#   pts_tot = 0
#   inicio = input('Bem vindo ao jogo de acerte o numero! Nos vamos escolher um numero aleatório de 0 a 10, e voce tem 5 tentativas para adivinhar o número. Vamos la? Aperte qualquer chave para comecar. ')
#   while i != 6:
#     welcome = input(f'Digite sua {i+1}a tentativa: ')
#     if welcome.isnumeric() == False:
#       print('Opcao invalida. Por favor digite um numero entre 0 e 10')
#       continue
#     else:
#       trial = int(welcome)
#       if trial not in range(0,11):
#         print('O numero que voce digitou esta fora do intervalo entre 0 e 10. Vamos te dar mais uma chance')
#         continue
#       elif trial == num:
#         print(f'Parabens! Voce acertou o numero na sua {i+1}a tentativa e ganhou {pts[i]} pontos nessa rodada!')
#         pts_tot += pts[i]
#         print(f'A sua quantidade de pontos total ate agora e {pts_tot}.')
#         i,num = new_game(pts_tot)
#       else:
#         tip(trial,num)
#         i += 1
#       if i == 5:
#         print(f'Infelizmente nao foi dessa vez! O numero que escolhemos foi {num}. Voce ganhou {pts[i]} pontos nessa rodada.')
#         pts_tot += pts[i]
#         print(f'A sua quantidade de pontos total ate agora e {pts_tot}.')
#         i,num = new_game(pts_tot)
    

# num = r.randrange(0,10)
# print(num)
# start_game(num)




  
    


```
