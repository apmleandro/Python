```python
#1 Faça um Programa que peça dois números e imprima o maior deles.

num1=int(input('Digite um numero: '))
num2=int(input('Digite um numero: '))

if num1 == num2:
  print('Os numeros sao iguais! Digite valores diferentes')
elif num1>num2:
  print(f'O numero maior e {num1}')
else:
  print(f'O numero maior e {num2}')

#2 Faça um Programa que peça um valor e mostre na tela se o valor é positivo ou negativo.

valor= int(input('Digite um valor: '))
if valor == 0:
  print('O valor digitado foi zero. Zero nao e nem positivo e nem negativo. E um numero neutro')
elif valor > 0:
  print('O numero que voce digitou e postivo')
else:
  print('O numero que voce digitou e negativo')

#3 Faça um Programa que verifique se uma letra digitada é "F" ou "M". Conforme a letra escrever: F - Feminino, M - Masculino, Sexo Inválido.

sexo = input('Porfavor, nos diga seu sexo. Digite F para feminino ou M para masculino em letra maiuscula ')
if sexo != 'F' and sexo != 'M':
  print('Sexo invalido')
elif sexo == 'F':
  print('F - Feminino')
else:
  print('M - Masculino')

#4 Faça um Programa que verifique se uma letra digitada é vogal ou consoante.

letra=input('Digite uma letra : ')
if letra in ('a','e','i','o','u') or letra in ('A','E','I','O','U') :
  print('A letra digitada e uma vogal')
else:
  print('A letra digitada e uma consoante')
  
#5:
# Faça um programa que pergunte o preço de três produtos e informe qual produto você deve comprar, sabendo que a decisão é sempre pelo mais barato.

menor=10000000000000
for i in range(3):
  cost_prod = float(input(f'Qual o preco do produto {i+1}? '))
  if cost_prod < menor:
    menor = cost_prod
    n_prod = i+1

print(f'O produto que irei comprar sera o produto numero {n_prod}, porque ele e o mais barato e custa {menor} reais')



#6:
# Um posto está vendendo combustíveis com a seguinte tabela de descontos:
# Álcool:
# até 20 litros, desconto de 3% por litro
# acima de 20 litros, desconto de 5% por litro
# Gasolina:
# até 20 litros, desconto de 4% por litro
# acima de 20 litros, desconto de 6% por litro

# Escreva um algoritmo que leia o número de litros vendidos, o tipo de combustível (codificado da seguinte forma: A-álcool, G-gasolina), calcule e imprima o valor a ser pago pelo cliente sabendo-se que o preço do litro da gasolina é R$ 2,50 o preço do litro do álcool é R$ 1,90.


preco_gas_litro = 2.50
preco_alc_litro = 1.90

litro = float(input('Quantos litros de combustivel foram vendidos? '))
combustivel = input('''Digite a letra maiscula correspondente ao combustivel utilizado?
A-álcool
G-gasolina
''')
if combustivel != 'A' and combustivel != 'G' :
  print('Opcao invalida')
elif combustivel == 'A' and litro <= 20:
  preco = preco_alc_litro * litro * (1-0.03)
elif combustivel == 'A' and litro > 20:
  preco = preco_alc_litro * litro * (1-0.05)
elif combustivel == 'G' and litro <= 20:
  preco = preco_gas_litro * litro * (1-0.04)
elif combustivel == 'G' and litro > 20:
  preco = preco_gas_litro * litro * (1-0.06)

print(f'O preco a ser pago pelo cliente e {preco}.')
  

#7:
# Faça um Programa que peça os 3 lados de um triângulo. O programa deverá informar se os valores podem ser um triângulo. Indique, caso os lados formem um triângulo, se o mesmo é: equilátero, isósceles ou escaleno.
# Dicas:
# Três lados formam um triângulo quando a soma de quaisquer dois lados for maior que o terceiro;
# Triângulo Equilátero: três lados iguais;
# Triângulo Isósceles: quaisquer dois lados iguais;
# Triângulo Escaleno: três lados diferentes;

lado1 = float(input('Qual o comprimento do primeiro lado do triangulo? '))
lado2 = float(input('Qual o comprimento do segundo lado do triangulo? '))
lado3 = float(input('Qual o comprimento do terceiro lado do triangulo? '))

sum1 = (lado1 + lado2)
sum2 = (lado2 + lado3)
sum3 = (lado1 + lado3)

if sum1 > lado3 or sum2 > lado1 or sum3 > lado2:
  print('Estes valores podem ser um triangulo')
  if lado1 == lado2 == lado3:
    print('O triangulo que esses lados formariam seria Equilatero.')
  elif lado1 != lado2 != lado3:
    print('O triangulo que esses lados formariam seria Escaleno.')
  else:
    print('O triangulo que esses lados formariam seria Isoceles.')

```
